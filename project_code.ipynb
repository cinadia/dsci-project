```r
library(tidyverse)
library(repr)
library(readxl)
user_knowlage_data <- read_excel("Data_User_Modeling_Dataset_Hamdi Tolga KAHRAMAN.xls", sheet = 2) %>%
    select(STG:UNS) %>%
    mutate(UNS = as_factor(UNS))
head(user_knowlage_data)
ggplot(user_knowlage_data, aes(x = PEG, y = STG, colour = UNS)) +
    geom_point() +
    labs(x = "Exam Performance in Target Area of Knowlage", y = "Study Time for Target Area of Knowlage", color = "Understanding") +
    ggtitle("Exam Performance in Target Area of Knowlage VS \nStudy Time for Target Area of Knowlage")
ggplot(user_knowlage_data, aes(x = PEG, y = STG, colour = UNS)) +
    geom_point() +
    labs(x = "Exam Performance in Target Area of Knowlage", y = "Study Time for Prerequisites", color = "Understanding") +
    ggtitle("Exam Performance in Target Area of Knowlage VS \nStudy Time for Prerequisites")
ggplot(user_knowlage_data, aes(x = PEG, y = LPR, colour = UNS)) +
    geom_point() +
    labs(x = "Exam Performance in Target Area of Knowlage", y = "Exam Performance in Prerequisites", color = "Understanding") +
    ggtitle("Exam Performance in Target Area of Knowlage VS \nExam Performance in Prerequisites")
```
